3 macam visualisasi data

1. - pit.pie digunakan untuk membuat diagram seperti dibawah
 - dengan dimasukkannya plt.axis equal agar berbentuk bulat

```python
import matplotlib.pyplot as plt

#data_contoh
hari = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7']
score = [25, 30, 60, 50, 68, 90, 99]

#grafiknya
plt.figure(figsize=(4,4))
plt.pie(score, labels=hari, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart statistik osu')
plt.axis('equal')
plt.show()
```

2. - plt.grid digunakan untuk mendefiniskan suatu bentuk dari diagram
- color memiliki kode warna


```python
import matplotlib.pyplot as plt

#data_contoh
hari = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'H8', 'H9', 'H10']
jumlah_penjualan = [25, 50, 24, 21, 68, 20, 40, 80, 21, 14]

#grafikknya
plt.figure(figsize=(8,4))
plt.plot(hari, jumlah_penjualan, color='r')
plt.title('jualan ratu')
plt.xlabel('hari')
plt.ylabel('jumlah penjualan')
plt.grid(True)
plt.show()
```
